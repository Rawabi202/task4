```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
# creating dataframe
dataFrame = pd.DataFrame({
    "price": ['24.49', '26.00','50.76','44.80'],"pages": [272.0,301.0 ,446.0,347.0]
})

# plot a Pie Chart for Registration Price column with label Car column
plt.pie(dataFrame["pages"], labels = dataFrame["price"])
plt.show()

```

```python
import matplotlib.pyplot as plot
# Earnings data for 4 quarters as a Python Dictionary

earningsData = { "price": ['24.49', '26.00','50.76','44.80'],"pages": [272.0,301.0 ,446.0,347.0] };

             

df = pd.DataFrame(data=earningsData);

 

# Draw a line chart

df.plot.line(title="price pages");

plot.show(block=True);
```
